```python
routes_filepath_list = ["D:\\Dokumente\\sumo\\maloja\\01_schools\\persontrips_maloja_school_resolved_offseason_taxi_modded_with_parents.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\02_work\\persontrips_maloja_work_resolved_offseason_taxi.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping_resolved_offseason_taxi.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\05_seniors\\persontrips_maloja_seniors_resolved_offseason_taxi.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\09_daytrips\\persontrips_maloja_daytrips_resolved_offseason_taxi.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\11_hotel\\persontrips_maloja_hotel_resolved_offseason_taxi.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\12_non_hotel\\persontrips_maloja_non_hotel_resolved_offseason_taxi.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\13_private\\persontrips_maloja_private_resolved_offseason_taxi.rou.xml"]
```

```python
tripinfos_file_xmltree = ET.parse(tripinfos_filepath)
tripinfos_file_xmlroot = tripinfos_file_xmltree.getroot()
```

```python
routes = []
```

```python
for routes_filepath in routes_filepath_list:
    routes_file_xmltree = ET.parse(routes_filepath)
    routes_file_xmlroot = routes_file_xmltree.getroot()
    routes = routes + list(routes_file_xmlroot)
```

```python
vehicles = []
```

```python
edges = []
```

```python
for trip in tripinfos_file_xmlroot:
    for segment in trip:
        if segment.tag == "ride":
            vehicle = segment.get("vehicle")
            vehicles.append(vehicle)
            if vehicle == "NULL":
                for route in routes:
                    if route.get("id") == trip.get("id"):
                        for part in route:
                            if part.tag == "ride":
                                edges.append(part.get("from"))
```

```python
results = {}

for unique_edge in set(edges):
    c = 0
    for edge in edges:
        if edge == unique_edge:
            c+=1
    if str(c) not in results:
        results[str(c)] = []
    results[str(c)].append(unique_edge)

dict(sorted(results.items()))
```
