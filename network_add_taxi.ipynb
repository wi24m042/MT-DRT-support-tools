```python
import sumolib
import xml.etree.ElementTree as ET
```

```python
network_filepath = "osm.net.xml"
output_filepath = "osm_modded.net.xml"
```

```python
network_file_sumolib = sumolib.net.readNet(network_filepath)
```

```python
network_file_xmltree = ET.parse(network_filepath)
network_file_xmlroot = network_file_xmltree.getroot()
```

```python
bus_not_taxi = []

for edge in network_file_sumolib.getEdges():
    edge_id = edge.getID()
    bus = edge.allows("bus")
    taxi = edge.allows("taxi")
    if(bus == True and taxi == False):
        bus_not_taxi.append(edge_id)

print("Number of edges which allow buses but not taxies/DRT vehicles: " + str(len(bus_not_taxi)))
```

```python
edge_counter = 0
lane_counter = 0

for net_child in network_file_xmlroot:
    if net_child.tag == "edge" and net_child.get("id") in bus_not_taxi:
        for edge_child in net_child:
            edge_child.set("allow", str(edge_child.get("allow")) + " taxi")
            lane_counter += 1
        edge_counter += 1

print("Number of edges on which taxis/DRT vehicles were added: " + str(edge_counter))
print("Number of lanes on which taxis/DRT vehicles were added: " + str(lane_counter))
```

```python
network_file_xmltree.write(output_filepath)
```
