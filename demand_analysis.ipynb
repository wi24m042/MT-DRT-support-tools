```python
routes_filepath_list = ["D:\\Dokumente\\sumo\\maloja\\09_daytrips\\persontrips_maloja_daytrips_resolved_offseason_taxi_w15.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\11_hotel\\persontrips_maloja_hotel_resolved_offseason_taxi_w15.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\12_non_hotel\\persontrips_maloja_non_hotel_resolved_offseason_taxi_w15.rou.xml",
                        "D:\\Dokumente\\sumo\\maloja\\13_private\\persontrips_maloja_private_resolved_offseason_taxi_w15.rou.xml"]
```

```python
for routes_filepath in routes_filepath_list:

    routes_file_xmltree = ET.parse(routes_filepath)
    routes_file_xmlroot = routes_file_xmltree.getroot()

    print(routes_filepath)

    for person_child in list(routes_file_xmlroot):
        vehicle = False
        for leg in person_child:
            if leg.tag == "ride":
                vehicle = True

        if vehicle == True:
            if str(person_child.get("id"))[0:2] == "se":
                trips_dict["type"].append("senior")
            elif str(person_child.get("id"))[0:1] == "s":
                trips_dict["type"].append("student")
            elif str(person_child.get("id"))[0:1] == "p":
                trips_dict["type"].append("guardian")
            elif str(person_child.get("id"))[0:1] == "w":
                trips_dict["type"].append("working")
            elif str(person_child.get("id"))[0:1] == "g":
                trips_dict["type"].append("shopping")
            elif str(person_child.get("id"))[0:1] == "d":
                trips_dict["type"].append("daytrip")
            elif str(person_child.get("id"))[0:2] == "th":
                trips_dict["type"].append("hotel")
            elif str(person_child.get("id"))[0:2] == "tn":
                trips_dict["type"].append("non-hotel")
            elif str(person_child.get("id"))[0:2] == "tp":
                trips_dict["type"].append("private")
        
            trips_dict["time"].append(float(str(person_child.get("depart"))))
            humantime = float(str(person_child.get("depart")))/60/60
            hour = str(int(humantime))
            minutes = str((int((humantime-int(humantime))*60)))
            minute_filler = "0" if int(minutes) < 9 else ""
            trips_dict["humantime"].append(hour + ":" + minute_filler + minutes)
```

```python
hours = np.arange(6, 21, 1)
```

```python
hourly_count = []
for hour in hours:
    hourly_count.append(len([i for i in trips_dict["time"] if hour*60*60 <= i <= (hour+1)*60*60]))

hourly_count
```
