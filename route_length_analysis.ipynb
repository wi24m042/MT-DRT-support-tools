```python
import xml.etree.ElementTree as ET
```

```python
#tripinfos_filepath = "D:\\Dokumente\\sumo\\2026-03-29-23-23-13\\14_drt_solving\\split\\tourists\\96_CH_ES_50_15\\tripinfos_drt.xml"
tripinfos_filepath = "D:\\Dokumente\\sumo\\2026-03-29-23-23-13\\14_drt_solving\\split\\locals\\2v164_50_5\\tripinfos_drt.xml"
```

```python
tripinfos_file_xmltree = ET.parse(tripinfos_filepath)
tripinfos_file_xmlroot = tripinfos_file_xmltree.getroot()
```

```python
results = {}

for trip in tripinfos_file_xmlroot:
    if trip.tag == "tripinfo":
        if trip.get("routeLength") not in results:
            results[trip.get("routeLength")] = []
        results[trip.get("routeLength")].append(trip.get("id"))

dict(sorted(results.items()))
```

```python
minimum = (1000000000,"")
maximum = (0,"")
over400km = 0
under1km = 0
under10km = 0

for trip in tripinfos_file_xmlroot:
    if trip.tag == "tripinfo":
        if minimum[0] > float(str(trip.get("routeLength"))):
            minimum = (float(str(trip.get("routeLength"))), trip.get("id"))

        if maximum[0] < float(str(trip.get("routeLength"))):
            maximum = (float(str(trip.get("routeLength"))), trip.get("id"))

        if float(str(trip.get("routeLength"))) > 386000:
            over400km+=1

        if float(str(trip.get("routeLength"))) < 1000:
            under1km+=1

        if float(str(trip.get("routeLength"))) < 10000:
            under10km+=1
```
