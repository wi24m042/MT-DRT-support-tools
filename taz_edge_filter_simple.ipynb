```python
import sumolib
import xml.etree.ElementTree as ET
```

```python
network_filepath = "simple.net.xml"
taz_filepath = "taz_maloja.add.xml"
output_filepath = "taz_maloja_simple.add.xml"
```

```python
net = sumolib.net.readNet(network_filepath)
```

```python
taz_file_xmltree = ET.parse(taz_filepath)
taz_file_xmlroot = taz_file_xmltree.getroot()
```

```python
changes = 1
while changes > 0:
    changes = 0
    for taz in taz_file_xmlroot:
        for taz_sink_source in taz:
            edge_id = taz_sink_source.get("id")
            taxi = False
            try:
                taxi = net.getEdge(edge_id).allows("taxi")
            except KeyError:
                pass
            if taxi == False:
                taz.remove(taz_sink_source)
                changes+=1

    print("Number of TAZ sources/sinks that were removed: " + str(changes))
```

```python
changes = 1
while changes > 0:
    changes = 0
    for taz in taz_file_xmlroot:
        for taz_sink_source in taz:
            edge_id = taz_sink_source.get("id")
            pedestrian = False
            try:
                pedestrian = net.getEdge(edge_id).allows("pedestrian")
            except KeyError:
                pass
            if pedestrian == False:
                taz.remove(taz_sink_source)
                changes+=1

    print("Number of TAZ sources/sinks that were removed: " + str(changes))
```

```python
changes = 1
while changes > 0:
    changes = 0
    for taz in taz_file_xmlroot:
        for taz_sink_source in taz:
            taz_sink_source_id = taz_sink_source.get("id")
            edge = net.getEdge(taz_sink_source_id)
            outgoings = edge.getOutgoing()
            if(len(outgoings) == 1):
                for outgoing in outgoings:
                    outgoing_outgoings = outgoing.getOutgoing()
                    if (len(outgoing_outgoings) == 1):
                        for outgoing_outgoing in outgoing_outgoings:
                            if(outgoing_outgoing.getID() == edge.getID()):
                                taz.remove(taz_sink_source)
                                changes+=1
                                
    print("Number of TAZ sources/sinks that were removed: " + str(changes))
```

```python
taz_file_xmltree.write(output_filepath)
```
