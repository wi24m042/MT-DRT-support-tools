```python
import xml.etree.ElementTree as ET
```

```python
persontrips_filepath = "persontrips_maloja_work_resolved_offseason_taxi_w15.rou.xml"
output_filepath = "persontrips_maloja_work_resolved_offseason_taxi_w15.rou.xml"
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
for person_child in list(persontrips_file_xmlroot):
    person_child.set("id", "w" + str(person_child.get("id")))
```

```python
persontrips_file_xmltree.write(output_filepath, encoding="utf-8")
```
