```python
import xml.etree.ElementTree as ET
```

```python
tripinfos_filepath = "D:\\Dokumente\\sumo\\maloja\\14_drt_solving\\split\\tourists\\96_CH_ES_50_15\\stopinfos_drt.xml"
#tripinfos_filepath = "D:\\Dokumente\\sumo\\maloja\\14_drt_solving\\split\\locals\\164_50_5\\stopinfos_drt.xml"
```

```python
tripinfos_file_xmltree = ET.parse(tripinfos_filepath)
tripinfos_file_xmlroot = tripinfos_file_xmltree.getroot()
```

```python
vehicle_count = 96
```

```python
vehicles = {}
loadedPersons = 0

for stop in tripinfos_file_xmlroot:
    if stop.get("id") not in vehicles:
        vehicles[stop.get("id")] = 0
    vehicles[stop.get("id")] += int(str(stop.get("loadedPersons")))
    loadedPersons += int(str(stop.get("loadedPersons")))

loadedPersons/vehicle_count
```
