```python
import xml.etree.ElementTree as ET
import copy
```

```python
persontrips_filepath = "persontrips_maloja_work.xml"
persontrips_filepath_output = "persontrips_maloja_work_modded.xml"
routes_filepath = "persontrips_maloja_work_resolved_offseason_bus.rou.xml"
routes_filepath_output = "persontrips_maloja_work_resolved_offseason_bus_modded.rou.xml"
```

```python
demand = {
    "taz_St_Moritz-taz_Silvaplana": 35,
    "taz_Silvaplana-taz_St_Moritz": 36,
    "taz_Silvaplana-taz_Sils_Engadin": 5,
    "taz_Sils_Engadin-taz_Silvaplana": 2,

    "taz_Sils_Engadin-taz_Bregaglia": 0,
    "taz_Bregaglia-taz_Sils_Engadin": 11,
    "taz_St_Moritz-taz_Sils_Engadin": 13,
    "taz_Sils_Engadin-taz_St_Moritz": 12,

    "taz_St_Moritz-taz_Bregaglia": 3,
    "taz_Bregaglia-taz_St_Moritz": 19,
    "taz_Silvaplana-taz_Bregaglia": 0,    
    "taz_Bregaglia-taz_Silvaplana": 4,

    "taz_St_Moritz-taz_St_Moritz": 192,
    "taz_Silvaplana-taz_Silvaplana": 7,
    "taz_Sils_Engadin-taz_Sils_Engadin": 8,
    "taz_Bregaglia-taz_Bregaglia": 41,

    "taz_St_Moritz_Sink-taz_St_Moritz": 125,
    "taz_St_Moritz_Sink-taz_Silvaplana": 33,
    "taz_St_Moritz_Sink-taz_Sils_Engadin": 7,
    "taz_St_Moritz_Sink-taz_Bregaglia": 1,

    "taz_St_Moritz-taz_St_Moritz_Sink": 61,
    "taz_Silvaplana-taz_St_Moritz_Sink": 18,
    "taz_Sils_Engadin-taz_St_Moritz_Sink": 20,
    "taz_Bregaglia-taz_St_Moritz_Sink": 22
}
demand
```

```python
demand_to_work = copy.deepcopy(demand)
demand_from_work = copy.deepcopy(demand)
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
routes_file_xmltree = ET.parse(routes_filepath)
routes_file_xmlroot = routes_file_xmltree.getroot()
```

```python
for trip_child in list(routes_file_xmlroot):
    has_ride = False
    for stage_child in list(trip_child):
        if stage_child.tag == "ride":
            if "intended" in stage_child.attrib:
                has_ride = True
                break

    if has_ride == True:
        if 21600 <= float(str(trip_child.get("depart"))) <= 36000:
            for persontrip_child in list(persontrips_file_xmlroot):
                if trip_child.get("id") == persontrip_child.get("id"):
                    demand_key = str(persontrip_child[0].get("fromTaz")) + "-" + str(persontrip_child[0].get("toTaz"))
                    print(demand_key)
                    if demand_to_work[demand_key] > 0:
                        demand_to_work[demand_key] = demand_to_work[demand_key] - 1
                        break
                    else:
                        routes_file_xmlroot.remove(trip_child)
                        break

        elif 54000 <= float(str(trip_child.get("depart"))) <= 72000:
            for persontrip_child in list(persontrips_file_xmlroot):
                if trip_child.get("id") == persontrip_child.get("id"):
                    demand_key = str(persontrip_child[0].get("toTaz")) + "-" + str(persontrip_child[0].get("fromTaz"))
                    
                    if demand_from_work[demand_key] > 0:
                        demand_from_work[demand_key] = demand_from_work[demand_key] - 1
                        break
                    else:
                        routes_file_xmlroot.remove(trip_child)
                        break

    if has_ride == False:
        routes_file_xmlroot.remove(trip_child)
```

```python
for persontrip_child in list(persontrips_file_xmlroot):
    persontrip_in_routes = False
    for trip_child in list(routes_file_xmlroot):
        if trip_child.get("id") == persontrip_child.get("id"):
            persontrip_in_routes = True
            break
    
    if persontrip_in_routes == False:
        persontrips_file_xmlroot.remove(persontrip_child)
```

```python
persontrips_file_xmltree.write(persontrips_filepath_output)
routes_file_xmltree.write(routes_filepath_output)
```
