```python
#list of lanes where DRT vehicles should spawn
vstops = ["531892740#1_0", "-210208895_0", "-543256264#0_0", "-1374129938_0", "-1083273106_0", "25418591#0_0", "-1068941219#6_0", "1426892794#0_0", "256959704#2_0"]
#number of DRT vehicles
fleet_size = 123
#output filepath
filepath_output = "drt_fleet.xml"
```

```python
fleet = []
vehicle_type_definition = \
"""    <vType id="taxi" vClass="taxi" personCapacity="4" color="magenta">
        <param key="has.taxi.device" value="true"/>
        <param key="device.taxi.stands-rerouter" value="rr0"/>
        <param key="device.taxi.end" value="86400"/>
    </vType>"""

#<param key="device.taxi.idle-algorithm" value="taxistand"/>
```

```python
for i in range(fleet_size):
    vehicle_definition = \
    """    <trip id="taxi{id}" depart="20520.00" type="taxi" >
        <stop lane="{lane_id}" triggered="person"/>
    </trip>""".format(id=i, lane_id=vstops[i%len(vstops)])
    fleet.append(vehicle_definition)
```

```python
with open(filepath_output, "w") as f:
    f.write(("<routes>\n"))
    f.write((vehicle_type_definition + "\n"))
    for vehicle in fleet:
        f.write((vehicle + "\n"))
    f.write(("</routes>"))
```
